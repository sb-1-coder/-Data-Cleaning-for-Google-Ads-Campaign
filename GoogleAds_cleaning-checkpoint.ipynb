```python
import pandas as pd
df = pd.read_csv("GoogleAds_uncleaned1.csv", encoding="latin1")
# 1. View the dataset to see with what kind of mistakes we're working with
df.info
```

```python
# 2. This table contains wrong format, writing of words, blank spaces. Let's start cleaning!
```

```python
df.columns
```

```python
# 3. Fixing the information in the columns
df["Campaign_Name"].unique()
```

```python
df["Campaign_Name"] = df["Campaign_Name"].replace({
    "DataAnalyticsCourse": "Data Analytics Course",
    "Data Analytics Corse": "Data Analytics Course",
    "Data Analytcis Course": "Data Analytics Course",
    "Data Anlytics Corse": "Data Analytics Course"
    
})
```

```python
df["Location"].unique()
```

```python
df["Location"] = df["Location"].replace ({
    "hyderabad" : "Hyderabad", 
    "Hyderbad" : "Hyderabad" , 
    "hydrebad" : "Hyderabad", 
    "HYDERABAD" : "Hyderabad"
})
```

```python
df["Device"].unique()
```

```python
df["Device"] = df["Device"].replace ({
    "desktop" : "Desktop", 
    "DESKTOP" : "Desktop" , 
    "tablet" : "Tablet", 
    "TABLET" : "Tablet",
    "MOBILE" : "Mobile",
    "mobile" : "Mobile"

})
```

```python
# 4. Viewing if the info is right
df.info
```

```python
# 5. Fixing the date format
```

```python
df["Ad_Date"].value_counts()
```

```python
df["Ad_Date"] = pd.to_datetime(df["Ad_Date"], format="mixed", dayfirst=True)
```

```python
df["Ad_Date"].value_counts()
```

```python
# 6. Counting missing values in each column.
df.isnull().sum()
```

```python
# 7. Filling the missing values
df.fillna("N/A", inplace=True)
```

```python
df.info
```

```python
# 8. Why we do not edit the keywords? Keywords are used to find the campaign's course, no matter how they are written, you'll find the info
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
# 9. The dataset is cleaned!
```

```python
# 10. Exporting the cleaned dataset
df.to_csv("dataset_limpio.csv", index=False)
```
